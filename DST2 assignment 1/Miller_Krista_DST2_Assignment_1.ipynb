d) format all the column names to lowercase and include underscore between column names that consist of two words.  for example, meanTemp should become mean_temp, and Max24hrPrep can become max_24hr_prep, and HighTemp becomes high_temp, etc..
reassign the dataframe with the formatted columns to the same variable, clean_boston_data.  print or output the columns of the clean_boston_data dataframe. 

```python
import re

def camel_to_snake(col_name):
    col_name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', col_name)
    col_name = re.sub('(.)([0-9]+)', r'\1_\2', col_name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', col_name).lower()


result = map(camel_to_snake, clean_boston_data.columns)
new_cols = list(result)

clean_boston_data.columns = new_cols

print(clean_boston_data.columns)

display(clean_boston_data.head(5))
```

d) using the boston_1 dataframe in question 1a, select the LowTemp, HighTemp, WarmestMin, ColdestHigh, AvgMin, AveMax columns and use the 'pipeline' functionality in sklearn to transform the selected data using the MinMaxScalar() and Imputer() classes.  With the Imputer() class, the missing data in each column should be imputed using the mean value for the column.  Assign the resulting DataFrame to a variable called pipeline_data.  Print the pipeline_data.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

boston_trim = boston_1[["LowTemp", "HighTemp", "WarmestMin", "ColdestHigh", "AveMin", "AveMax"]]


imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
minmax = MinMaxScaler()
pipe = Pipeline([("impute", imp_mean), ("scaler", minmax)])

pipeline_data = pipe.fit_transform(boston_trim)

pipeline_data = (pd.DataFrame(pipeline_data, columns = ["LowTemp", "HighTemp", "WarmestMin", "ColdestHigh", "AveMin", "AveMax"]))
display(pipeline_data)
```

e) output the descriptive statistics of the pipeline_data including mean, median, variance, minimum value, maximum value, variance, standard deviation, and skewness.  you can use the apply() functionality of the pandas dataframe.
